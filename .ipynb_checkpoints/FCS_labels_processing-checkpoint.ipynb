## Post-processing of labels
### If combination of folder and file is not unique, take the last one

```python
labels = data_lb.copy()
```

```python
a = data_lb.duplicated(subset="Path", keep='last')    #Return boolean Series denoting duplicate rows
```

```python
labels = labels[~a]
```

```python
# Check lenght: labels correspond to data, without repetitions

#len(labels) == len(data_fcs)
```

```python
labels = labels.reset_index()
labels = labels.drop('index', axis = 1)
```

```python
np.count_nonzero(labels['Label'])
```

```python
# Fraction of anomalies

print('Fraction of anomalies:', np.count_nonzero(labels['Label'])/len(labels))
```
